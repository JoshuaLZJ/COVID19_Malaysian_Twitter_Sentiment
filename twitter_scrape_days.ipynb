```python
def clean_tweets(twitter):
  #cleaning up text
  #can rerun this multiple times

  for t in twitter:
    twitter[t] = twitter[t].replace('\n', '')
    
    #remove links
    
    hit = twitter[t].find('https')

    for i in range(hit+5, len(twitter[t])):
      if i == len(twitter[t]) - 1:
        twitter[t] = twitter[t][:hit]
        break
      else:
        if twitter[t][i] != ' ':
          continue
        else:
          twitter[t] = twitter[t][:hit] + twitter[t][i:]
          break

    #remove hashtags

    hit = twitter[t].find('#')

    for i in range(hit+5, len(twitter[t])):
      if i == len(twitter[t]) - 1:
        twitter[t] = twitter[t][:hit]
        break
      else:
        if twitter[t][i] != ' ':
          continue
        else:
          twitter[t] = twitter[t][:hit] + twitter[t][i:]
          break

  return(twitter)
```

```python
#processing all tweets

days = 574
now = datetime.datetime(2021, 10, 11)
date_lst = []
single = datetime.timedelta(1)

for i in range(days):
  date_lst.append(str(now)[:10])
  now -= single

date_lst.append('2020-03-16')

for d in range(len(date_lst)-1):
  filename = '/content/drive/MyDrive/Honours/Twitter/any_language/574_days/' + date_lst[d] + '_text-query-tweets_malay_NRP.json'
  tweets_df = pd.read_json(filename, lines=True)
  only_tweets_df = tweets_df['content']
  #re-adjust for one day off ler
  only_tweets_file = f'/content/drive/MyDrive/Honours/Twitter/any_language/574_days_only_tweets/{date_lst[d+1]}_only_tweets.json'
  only_tweets_df.to_json(only_tweets_file)
  with open(only_tweets_file) as fopen:
    twitter = json.load(fopen)
  
  #clean twitter 3 times
  for i in range(3):
    twitter = clean_tweets(twitter)
  
  twitter_lst = []

  for t in twitter:
    twitter_lst.append(twitter[t])
  
  with open(f'/content/drive/MyDrive/Honours/Twitter/any_language/574_days_processed/{date_lst[d+1]}_processed_tweets.json', 'w') as outfile:
    json.dump(twitter_lst, outfile)
```
